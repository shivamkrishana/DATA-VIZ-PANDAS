```python
#HEXAGONAL BIN PLOTS
```

```python
#A 2D LATTICE is divide dinto a number of hexagonal grids of size n*m
#based on number of data points of the distribution of 2 variables x,y 
#that fall under any hexagon, the hexagons are given a given color
```

```python
#A hexagonal bin plot is really very helpful when number of points in 
#the distribution is very large
```

```python
import pandas as pd
import numpy as np
```

```python
#lets demonstrate
#this time we will create new dataframe coz we need a dataframe with 
#a lot of points
```

```python
df=pd.DataFrame(np.random.randn(1000,2),columns=["a","b"])
```

```python
df
```

```python
#we have created the dataframe ... lets plot 
df.plot.hexbin(x="a",y="b")
```

```python
#lets customize
```

```python
df.plot.hexbin(x="a",y="b",gridsize=35,cmap="Oranges")
```
