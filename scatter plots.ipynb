```python
#for plotting to scatter using DataFrame class, there is a member 
#called plot.
```

```python
#syntax:
```

```python
DataFrame.plot.scatter(x,y,s=none,c=none)
```

```python
s---->size of dots
c---->colour of dots
```

```python
#demonstrate
```

```python
import pandas as pd
```

```python
df1=pd.read_csv("df1",index_col=0)
```

```python
df1
```

```python
df1.plot.scatter(x="A",y="B")
```

```python
#WE CAN CHANGE THE COLORMAP
```

```python
df1.plot.scatter(x="A",y="B",c="C",cmap="coolwarm")
```

```python
#we can change size of dots to indicate value of a column
```

```python
#size---->takes a list
```

```python
df1.plot.scatter(x="A",y="B",s=df1["C"]*200,c="C",cmap="coolwarm")
```
